```python
# Import Libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading Excel File
df = pd.read_excel('C:\\Users\\DELL\\Desktop\\Financial_Statement_Analysis\\Tata_Steel_Cleaned_Financials.xlsx')
```

```python
# Display the Data
print(df.head())
```

```python
# plot Revenue VS Net Profit
plt.figure(figsize=(8,5))
plt.plot(df['Year'],df['Revenue (Cr)'], marker='o', label='Revenue')
plt.plot(df['Year'],df['Net Profit'], marker='o', label='Net Profit')
plt.title("Revenue vs Net Profit (Tata Steel)")
plt.xlabel("Year")
plt.ylabel("Crore")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot ROE vs ROCE
plt.figure(figsize=(8,5))
plt.plot(df['Year'],df['ROE (%)'], marker='s', label='ROE %')
plt.plot(df['Year'],df['ROCE (%)'], marker='s', label='ROCE %')
plt.title("ROE vs ROCE")
plt.xlabel("year")
plt.ylabel("Percentage")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Debt-Equity and P/E Ratio 
fig, ax1 = plt.subplots(figsize=(8,5))

ax1.set_xlabel('Year')
ax1.set_ylabel('Debt/Equity', color='tab:blue')
ax1.plot(df['Year'], df['Debt/Equity'], marker='x', color='tab:blue', label='Debt/Equity')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()  # secondary axis
ax2.set_ylabel('P/E Ratio', color='tab:red')
ax2.plot(df['Year'], df['P/E Ratio'], marker='x', color='tab:red', label='P/E Ratio')
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.title("Debt/Equity vs P/E Ratio")
plt.grid()
plt.show()
```

```python


plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['Debt/Equity'], marker='o', color='teal', linewidth=2)
plt.title("Tata Steel: Debt-to-Equity Ratio Over Years")
plt.xlabel("Year")
plt.ylabel("Debt/Equity Ratio")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['P/E Ratio'], marker='s', color='purple', linewidth=2)
plt.title("Tata Steel: Price to Earnings (P/E) Ratio Over Years")
plt.xlabel("Year")
plt.ylabel("P/E Ratio")
plt.grid(True)
plt.tight_layout()
plt.show()
```
